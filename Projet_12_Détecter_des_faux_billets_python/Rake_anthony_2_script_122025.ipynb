```python
import pandas as pd
import pickle
```

```python
with open("modele_logreg.pkl", "rb") as f:
    modele = pickle.load(f)

print("Modèle chargé avec succès.")
```

```python
df_prod = pd.read_csv("billets_production.csv", sep=",")
df_prod.head()
```

```python
# Colonnes attendues par le pipeline (celles vues pendant l'entraînement)
colonnes_modele = list(modele.feature_names_in_)
print("Colonnes attendues par le modèle :", colonnes_modele)

print("\nColonnes du fichier production :", list(df_prod.columns))
```

```python
df_prod_vf = df_prod.drop(columns=['id'])
print("\nColonnes du fichier production_vf :", list(df_prod_vf.columns))
```

```python
y_pred_modele = modele.predict(df_prod_vf)
print(y_pred_modele)
```

```python
resultat = pd.DataFrame([False, False, False, True, True], columns=['resultat'])
resultat
```

```python
pd.concat([df_prod, resultat], axis=1)
```

```python
def detect_faux_billets(dataset):
    df_prod_vf = dataset.drop(columns=['id'])
    y_pred_modele = modele.predict(df_prod_vf)
    resultat = pd.DataFrame(pd.Series(y_pred_modele), columns=['resultat'])
    resultat_vf = pd.concat([df_prod, resultat], axis=1)
    return resultat_vf
```

```python
detect_faux_billets(df_prod)
```
